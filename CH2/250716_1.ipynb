```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. 등속운동 (Constant Velocity Motion)

- 정의: 가속도가 0이고, 속도가 일정한 운동
- 방정식: 위치 $x(t) = x_0 + v \cdot t$

## 1차원

```python
# 2. 점 초기화
dt = 0.1        # 시간 간격 (s)
t_max = 5       # 최대 시간 (s)
steps = int(t_max / dt)

x_init = 0
vx_init = 1

x = np.zeros(steps)
y = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
  else:
    x[t] = x[t-1] + vx_init * dt
```

```python
t = np.linspace(0, t_max, steps)
plt.plot(t, x)
plt.xlabel('time')
plt.ylabel('position')
plt.show()
```

## 2차원

```python
# 2. 점 초기화
dt = 0.1        # 시간 간격 (s)
t_max = 5       # 최대 시간 (s)
steps = int(t_max / dt)

x_init = 0
y_init = 0
vx_init = 1
vy_init = 1

x = np.zeros(steps)
y = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    y[t] = y_init
  else:
    x[t] = x[t-1] + vx_init * dt
    y[t] = y[t-1] + vy_init * dt
```

```python
# 3D 라인 플롯 생성
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# 1차원 데이터를 사용하여 3D 라인 플롯 그리기
ax.plot(t, x, y)

# 축 라벨 설정
ax.set_xlabel('Time')
ax.set_ylabel('X Position')
ax.set_zlabel('Y Position')

# 플롯 표시
plt.show()
```

## 2. 등가속도운동 (Constant Acceleration Motion)
여기서는 자유낙하운동을 다룬다.

- 정의: 가속도가 일정한 운동
- 방정식: 위치 $x(t) = x_0 + v_0 t + \tfrac{1}{2} a t^2$

```python
# 1. 점 초기화
dt = 0.1        # 시간 간격 (s)
t_max = 5       # 최대 시간 (s)
steps = int(t_max / dt)

x_init = 0
y_init = 0
vy_init = 0
ay_init = -9.8

y = np.zeros(steps)
vy = np.zeros(steps)
for t in range(steps):
  if t == 0:
    y[t] = y_init
    vy[t] = vy_init
  else:
    y[t] = y[t-1] + vy[t-1] * dt
    vy[t] = vy[t-1] + ay_init * dt
```

```python
t = np.linspace(0, t_max, steps)
plt.figure(figsize=(10,3))

plt.subplot(121)
plt.plot(t, y)
plt.xlabel('time')
plt.ylabel('position')

plt.subplot(122)
plt.plot(t, vy)
plt.xlabel('time')
plt.ylabel('velocity')
plt.show()
```

## 수치적 계산의 문제

## 3. 포물선운동 (Projectile Motion)

- 정의: 중력만을 받으며 포물선을 그리는 운동
- 방정식:
  - $x(t) = v_0 \cos(\theta) \; t$
  - $y(t) = v_0 \sin(\theta) \; t - \tfrac{1}{2} g t^2$

```python
# 1. 점 초기화
dt = 0.1        # 시간 간격 (s)
t_max = 3       # 최대 시간 (s)
steps = int(t_max / dt)

x_init = 0
y_init = 0
vx_init = 10
vy_init = 10
ax_init = 0
ay_init = -9.8

x = np.zeros(steps)
y = np.zeros(steps)
vx = np.zeros(steps)
vy = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    y[t] = y_init
    vx[t] = vx_init
    vy[t] = vy_init
  else:
    x[t] = x[t-1] + vx[t-1] * dt
    y[t] = y[t-1] + vy[t-1] * dt
    vx[t] = vx[t-1] + ax_init * dt
    vy[t] = vy[t-1] + ay_init * dt
```

```python
t = np.linspace(0, t_max, steps)
plt.plot(x, y)
plt.xlabel('X position')
plt.ylabel('Y position')
plt.show()
```

### 각도에 따른 물체 운동

```python
x_lst = []
y_lst = []
v_theta_lst = [np.pi/6, np.pi/4, np.pi/3]

for v_theta in v_theta_lst:
  dt = 0.001        # 시간 간격 (s)
  t_max = 5       # 최대 시간 (s)
  steps = int(t_max / dt)

  x_init = 0
  y_init = 0
  v_init = 5
  vx_init = v_init * np.cos(v_theta)
  vy_init = v_init * np.sin(v_theta)
  ax_init = 0
  ay_init = -9.8

  x = np.zeros(steps)
  y = np.zeros(steps)
  vx = np.zeros(steps)
  vy = np.zeros(steps)
  for t in range(steps):
    if t == 0:
      x[t] = x_init
      y[t] = y_init
      vx[t] = vx_init
      vy[t] = vy_init
    else:
      x[t] = x[t-1] + vx[t-1] * dt
      y[t] = y[t-1] + vy[t-1] * dt
      vx[t] = vx[t-1] + ax_init * dt
      vy[t] = vy[t-1] + ay_init * dt
    # 땅에 닿으면 계산 중단
    if y[t] < 0:
        break

  x_lst.append(x)
  y_lst.append(y)
```

```python
angles = ['$\pi/6$', '$\pi/4$', '$\pi/3$']
colors = ['red','blue','black']
for i in range(3):
  plt.plot(x_lst[i], y_lst[i], color=colors[i], label=angles[i])
plt.xlim([0, max([max(i) for i in x_lst])])
plt.ylim([0, max([max(i) for i in y_lst])])
plt.legend()
plt.show()
```

## 4. 조화진동 (Simple Harmonic Motion)

- 정의: 복원력이 변위에 비례하는 운동
- 방정식: $x(t) = A \cos(\omega t)$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
```

```python
fig, ax = plt.subplots()

# 1. 그림 범위 설정
ax.set_xlim([-10, 10])

# 2. 점 초기화
T = 20
dt = 0.1
steps = int(T/dt)

m = 1.0   # 물체의 질량
k = 1.0   # 스프링 상수
x_init = 5
vx_init = 0
pos_init = ax.scatter(x_init, 0)

x = np.zeros(steps)
vx = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    vx[t] = vx_init
  else:
    a = -(k/m) * x[t-1]
    vx[t] = vx[t-1] + a * dt
    x[t] = x[t-1] + vx[t] * dt

# 3) 매 프레임마다 호출될 함수: 각 scatter를 새 위치(0, y[t])로 이동
def animate(t):
    pos_init.set_offsets((x[t], 0))
    return (pos_init,)

# 4) 애니메이션 객체 생성
ani = animation.FuncAnimation(
    fig,                  # 어느 Figure에 그릴지
    animate,              # 매 프레임마다 호출할 함수
    repeat=True,          # 끝나면 다시 처음부터 반복
    frames=steps,          # 총 프레임 수 = time 만큼
    interval=50           # 프레임 간 간격(밀리초)
)

# 5) 정적 그림 제거, jshtml로 표시
plt.close(fig)
rc('animation', html='jshtml')
ani
```

```python
tarray = np.linspace(0, T, steps)
plt.plot(tarray, x)
plt.show()
```

## 5. 감쇠진동 (Damped Oscillation)

- 정의: 마찰 등 저항으로 진폭이 서서히 감소하는 진동
- 방정식: $x(t) = A e^{-\gamma t} \cos(\omega_0 t)$

```python
fig, ax = plt.subplots()

# 1. 그림 범위 설정
ax.set_xlim([-10, 10])

# 2. 점 초기화
T = 20
dt = 0.1
steps = int(T/dt)

m = 1.0   # 물체의 질량
k = 1.0   # 스프링 상수
c = 0.2   # 저항 계수
x_init = 5
vx_init = 0
pos_init = ax.scatter(x_init, 0)

x = np.zeros(steps)
vx = np.zeros(steps)
for t in range(steps):
  if t == 0:
    x[t] = x_init
    vx[t] = vx_init
  else:
    a = -(k/m)*x[t-1] - (c/m)*vx[t-1]
    vx[t] = vx[t-1] + a*dt
    x[t] = x[t-1] + vx[t]*dt

# 3) 매 프레임마다 호출될 함수: 각 scatter를 새 위치(0, y[t])로 이동
def animate(t):
    pos_init.set_offsets((x[t], 0))
    return (pos_init,)

# 4) 애니메이션 객체 생성
ani = animation.FuncAnimation(
    fig,                  # 어느 Figure에 그릴지
    animate,              # 매 프레임마다 호출할 함수
    repeat=True,          # 끝나면 다시 처음부터 반복
    frames=steps,          # 총 프레임 수 = time 만큼
    interval=50           # 프레임 간 간격(밀리초)
)

# 5) 정적 그림 제거, jshtml로 표시
plt.close(fig)
rc('animation', html='jshtml')
ani
```
